```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib
import numpy as np
```

```python
df = pd.read_csv('all_players.csv')
```

```python
similarity_columns = [
    'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Acceleration', 'Sprint Speed',
    'Positioning', 'Finishing', 'Shot Power', 'Long Shots', 'Volleys', 'Penalties',
    'Vision', 'Crossing', 'Free Kick Accuracy', 'Short Passing', 'Long Passing',
    'Curve', 'Dribbling', 'Agility', 'Balance', 'Reactions', 'Ball Control',
    'Composure', 'Interceptions', 'Heading Accuracy', 'Def Awareness',
    'Standing Tackle', 'Sliding Tackle', 'Jumping', 'Stamina', 'Strength', 'Aggression'
]
```

```python
df_similarity = df.dropna(subset=similarity_columns).copy()
```

```python
player_info = df_similarity[['Name', 'OVR', 'Position', 'Team']].reset_index(drop=True)
```

```python
X_similarity = df_similarity[similarity_columns]

# --- Scale the data ---
# This is crucial to give all attributes equal importance in the distance calculation
similarity_scaler = StandardScaler()
scaled_attributes = similarity_scaler.fit_transform(X_similarity)
```

```python
joblib.dump(similarity_scaler, 'models/similarity_scaler.pkl')
np.save('models/scaled_attributes.npy', scaled_attributes)
player_info.to_csv('models/player_info.csv', index=False)

print("Similarity components have been created and saved successfully!")
```
